# Matplotlib Part1

Matplotlib is the 'grandfather' library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general
* similar to matlab

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org/

## Importing

```python
import matplotlib.pyplot as plt
import numpy as np
```

>%matplotlib.inline: show matplot graphs in jupyter lab / in other in envs use plt.show()

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

> There are two ways to control matplot 
* Functional 
* Object oriented

```python
#FUNCTIONAL
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show()
```

> Creating Multiplots 

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

> plt.figure() : create an imaginary canvas

```python
# Object Oriented Method
fig = plt.figure()

# each number is % of the total canvas size
#left, bottom, width, height 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
# move the axes to the bottom corner
# axes3 = fig.add_axes([0.5,0.1,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')


axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# Matplotlib Part2

> plt.subplots() : specify the number of rows and columns
> * In this case, axes is a list of axes objects  
> * plt.tightlay_out() take care of plots overlapping each other

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('Figure 1')

axes[1].plot(y,x)
axes[1].set_title('Figure 2')

#axes.plot(x,y)
# for current_ax in axes:
#     current_ax.plot(x,y)

plt.tight_layout()
```

## Figure SIze and DPI

```python
#dpi : pixels per inch
fig = plt.figure(figsize=(8,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2), dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

> fig.savefig({filename')Save figures to output files 

```python
fig.savefig('my_picture.png', dpi=200)
```

> ax.legend() : looks for labels 
* ax.plot(x,y,label='x sqaured')  
* ax.legend(loc=10) : locate the legend in different locations

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'y', label='x squared')
ax.plot(x,x**3, 'g', label='x cubed')

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.legend()
```

# Matplotlib Part3

> 'ax.plot()'
* RGB Hex Code could also work
* linewidth (or lw) : 1 is default 
* alpha : transparency
* linestyle (or ls) : '-.' '--' '-' 
* marker : 'o' '+' '*' '1' '2'
* markersize : size of the marker
* markerfacecolor : 'yellow'
* markeredgewidth : 3 

> `ax.set_xlim([{int},{int}])` : set limit values for the axis

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# RGB Hex Code could also work
# linewidth (or lw) : 1 is default 
# alpha : transparency
# linestyle (or ls) : '-.' '--' '-' 
# marker : 'o' '+' '*' '1' '2'
# markersize : size of the marker
ax.plot(x,y, color='purple', linewidth=2, linestyle='-.', marker='2') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
