# K-Means Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

### `make_blobs(n_samples=200, n_features=2, centers=4, cluster_std =1.8, random_state=101)` : make artificial data
* `n_samples` : number of data
* `n_features` : nubmer of features
* `centers` : number of clusters
* `cluster_std` : std of each cluster


```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std =1.8, random_state=101)
```

```python
data
```

```python
plt.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 8)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap='coolwarm')
```

### `plt.subplots(1,2, sharey=True, figsize=(10,6))`
* `sharey` : share the axis

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap='coolwarm')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='coolwarm')
```
