# Merging, Joining and Concatenating

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
pd.concat([df1,df2,df3])
```

> __concat(axis=1)__: concat by column

```python
pd.concat([df1,df2,df3], axis=1)
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

## Merging 
The __merge__ function allows you to merge DFs together using a similar login as merging SQL Tables together

```python
pd.merge(left,right,how='inner',on='key1')
```

```python
pd.merge(left,right,how='inner',on=['key1','key2'])
```

```python
pd.merge(left,right, how='outer', on=['key1','key2'])
```

```python
pd.concat([left,right],axis=1)
```

```python
pd.merge(left,right, how='right',on=['key1','key2'])
```

> The results below will be same with the results above

```python
pd.merge(right,left,how='left',on=['key1','key2'])
```

```python
pd.merge(left,right,how='left',on=['key1','key2'])
```

## Joining
Joining is a convenient method for combining the columns of two potentially differently-index DFs into a single result DF

*join based on the indexes (merge : based on columns)*

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, how='inner', left_index=True, right_index=True)
```

```python
left.join(right) # isn't is left outher join by default? 
```

```python
left.join(right, how='outer') 
```
