# Categorical Plots

```python
import seaborn as sns
%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

## sns.barplot(x={x column}, y={y column}, data={data})  : create a bar plot
> estimator : include aggregation functions. you can use your own functions

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

# sns.countplot(x={column}, data =tips) : counting the data

```python
sns.countplot(x='sex', data=tips)
```

# sns.boxplot(x={column}, y={column}, data={data}) : show box plots
> hue : split the box plots based on the column

```python
sns.boxplot(x='day', y='total_bill',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill',data=tips, hue='sex')
```

```python
sns.boxplot(x='tip', y='total_bill',data=tips)
```

# sns.violinplot(x={variable}, y={varible}, data={data} : shows the kde distribution of y  
> split=T/F : combine the plots if T

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

# sns.stripplot(x=,y=,data=) : scatter plot divided by x variables
> jitter : spreads the points stacked on top of each other  
> hue : add one more variable into consideration  
> split : split the plot when hue is applied

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

# sns.swarmplot(x=, y=, data=) : show scatter plot which stacks the plots based on their occurences
> * however they are computationally heavy  
> * also people might not be used to it

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

> These kinds of plots are suggested for exploratory uses

```python
sns.violinplot(x='day',y='total_bill', data=tips)
sns.swarmplot(x='day',y='total_bill', data=tips, color='black')
```

# sns.factorplot(x=,y=,data=, kind=) : can draw any kinds of plot depending on the 'kind' parameter
> kind : 'bar', 'violin', 'swarm' ...

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```
