# Grids

```python
import seaborn as sns 
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

### sns.PairGrid({data}) : create a canvas for pair plots 
 * map_upper({plot_class}) : display the mentioned plot in the upper part  
* map_lower({plot_class}) : display the mentioned plot in the lower part  
 * map_diag({plot_class}) : display the mentioned plot in the diagnoal part

```python
g = sns.PairGrid(iris)
g.map_upper(sns.scatterplot)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head(6)
```

### sns.FacetGrid(tips, col={column_name}, row={column_name}) : create a grid which contains the designated number of grids
* g.map({plot_class} , {column of the data you want to analyze}) : fill in the grid

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

### g.map(sns.scatterplot, {col1},{col2}) : include more column names if the plot requires more than one column

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.scatterplot, 'total_bill','tip')
```
