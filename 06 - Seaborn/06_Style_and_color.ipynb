# Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(6)
```

### `sns.set_style({style_name})`: 'darkgrid', 'white', 'whitegrid'

### `sns.despine()` : remove the edges of the plot square 

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
sns.despine()
```

### You can use matplotlib library to set the configuration for seaborn plots

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

### `sns.set_context({context_name}, fontsize={int})` : configure the size of the plot

```python
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('notebook')
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', palette='seismic')
```
