# Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### `sns.lmplot(x=, y=, data=)`
* `scatter_kws={dictionary parameters}` : input the underlying variables for drawing plots
* `col={column}` : divide the plot with a certain variable 
* `row={column}` : divide the plot with a certain variable 
* `aspect={float}` : ratio between height and width of the pltos
* `size={float}` : size of the plot 

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, height=13)
```
