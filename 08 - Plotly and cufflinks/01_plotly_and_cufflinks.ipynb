```python
# DATA 
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df
```

```python
df2 = pd.DataFrame({'Category' : ['A', 'B', 'C'], 'Values':[32,43,50]})
```

```python
df2
```

```python
df.plot()
```

### `df.iplot()` : create an interactive plot

```python
df.iplot()
```

### `df.iplot()`
* `kind='scatter'`
* `x='A', y='B'`
* `mode='markers'`
* `size=20`

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', size=20)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

### `df.sum().iplot()` : aggregation functions(count...) can be used

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y': [10,20,30,20,10], 'z': [500, 400, 300, 200, 100]})
```

```python
df3.iplot(kind='surface')
```

* `colorscale` : 'rdylbu', each letter stands for the initial letter of the color

```python
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
df['A'].iplot(kind='hist', bins=25)
```

```python
df.iplot(kind='hist', bins=25)
```

* `kind='spread'` show the spread between variables

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
```

### `df.scatter_matrix()` : scatter plots for all the variables
* every variable has to be numerical variable

```python
df.scatter_matrix()
```
